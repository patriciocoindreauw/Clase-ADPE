```python
#Agregar una columna igual a 1 si la columna gender es igual a 'M' y 0 si es 'F'
datos2['Genero']=datos4['Gender'].apply(lambda x: 1 if x=='M' else 0)

#Agregar una columna con el promedio de las columnas númericas
#datos2['Promedio']=datos4[numeric_columns].mean(axis=1)

# Asegurarse de que todas las columnas involucradas en operaciones aitméticas sean numéricas
numeric_columns=datos4.select_dtypes(include=['number']).columns
datos2['Promedio']=datos4[numeric_columns].mean(axis=1)

#Agregar una columna con la suma de las columnas númericas
datos2['Suma']=datos4[numeric_columns].sum(axis=1)

#Agregar una columna con la suma de dos columnas
datos2['Suma2']=datos4['Customer_Age']+datos4['Dependent_count']

#Agregar una columna con la resta de las columnas
datos2['Resta']=datos4['Customer_Age']-datos4['Dependent_count']

#Agregar una columna con la multiplicación de las columnas
datos2['Multiplicacion']=datos4['Customer_Age']*datos4['Dependent_count']

#Agregar una columna con la división de las columnas
datos2['Division']=datos4['Customer_Age']/datos4['Dependent_count']

#Mover la columna 'C' al inicio
columna=datos2.pop("Division") # Elimina la columna y la guarda en una variable
datos2.insert(0,"Division",columna) # Inserta la columna en la posición deseada

datos2.head()
```
