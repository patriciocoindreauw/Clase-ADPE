```python
# 3. ¿Cuál es el porcentaje (%) de observaciones que reportan ser de la región norte?
porcentaje_norte = (data['region'] == 'Norte').sum() / len(data) * 100
print(porcentaje_norte)
```

```python
# 4. ¿Cuál es la categoría más frecuente en la variable product_category?
# Respuesta: número de apariciones (frecuencia)
frecuencia_max = data['product_category'].value_counts().max()
print(frecuencia_max)
```

```python
# Alternativas para obtener el número de variables
len(data.columns)
# o simplemente
data.shape[1]
```

```python
# Eliminar los missing values (filas con valores faltantes)
data_sin_missing = data.dropna()

# Contar cuántas observaciones quedaron
num_observaciones = data_sin_missing.shape[0]
print(f"Observaciones después de eliminar missing values: {num_observaciones}")

# También puedes ver cuántas se eliminaron
print(f"Observaciones eliminadas: {data.shape[0] - data_sin_missing.shape[0]}")
```

```python
# Promedio del precio de las unidades (con 2 decimales)
promedio_precio = data_sin_missing['unit_price'].mean()
print(f"Promedio del precio de las unidades: {promedio_precio:.2f}")

# O si modificaste el DataFrame original:
# promedio_precio = data['unit_price'].mean()
# print(f"Promedio del precio de las unidades: {promedio_precio:.2f}")
```

```python
# Promedio de unidades vendidas (con 2 decimales)
promedio_unidades = data_sin_missing['units'].mean()
print(f"Promedio de unidades vendidas: {promedio_unidades:.2f}")

# O si modificaste el DataFrame original:
# promedio_unidades = data['units'].mean()
# print(f"Promedio de unidades vendidas: {promedio_unidades:.2f}")
```

```python
# Total de missing values en toda la base de datos
total_missing = data.isnull().sum().sum()
print(f"Total de missing values: {total_missing}")

# Si quieres ver los missing values por cada columna:
missing_por_columna = data.isnull().sum()
print("\nMissing values por columna:")
print(missing_por_columna)
```
